```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
data_complete = pd.merge(clinical_trial_data, mouse_drug_data, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
data_complete.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol_data = data_complete.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
df_mean_tumor_vol_data = pd.DataFrame(mean_tumor_vol_data)
df_mean_tumor_vol_data.reset_index(inplace=True)

# Preview DataFrame
df_mean_tumor_vol_data
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint 
from scipy import stats
std_tumor_vol_data = data_complete.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)']

# Convert to DataFrame
std_tumor_vol_data = pd.DataFrame(std_tumor_vol_data)
std_tumor_vol_data.reset_index(inplace=True)

std_tumor_vol_data.head()# Minor Data Munging to Re-Format the Data Frames
df_mean_tumor_vol_data_new = df_mean_tumor_vol_data.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
df_mean_tumor_vol_data_new.head()
```

```python
# Generate the Plot (with Error Bars) - Tumor Response to Treatment
x_axis = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
y_errors = 2
line_width = 1
label_Capomulin = "Capomulin"
fmt_Capomulin = "-ro" # red circle
label_Infubinol = "Infubinol"
fmt_Infubinol = "-b^" # blue triangle
label_Ketapril = "Ketapril"
fmt_Ketapril = "-gs" # green square
label_Placebo = "Placebo"
fmt_Placebo = "-kD" # black diamond


plt.errorbar(
    x_axis, # X
    df_mean_tumor_vol_data_new['Capomulin'],  # Y
    yerr = y_errors,     # Y-errors
    label = label_Capomulin,
    fmt = fmt_Capomulin,
    linewidth = 1 # width of plot line
    )

plt.errorbar(
    x_axis, # X
    df_mean_tumor_vol_data_new['Infubinol'],  # Y
    label = label_Infubinol,
    fmt = fmt_Infubinol
    )

plt.errorbar(
    x_axis, # X
    df_mean_tumor_vol_data_new['Ketapril'],  # Y
    label = label_Ketapril,
    fmt = fmt_Ketapril
    )

plt.errorbar(
    x_axis, # X
    df_mean_tumor_vol_data_new['Placebo'],  # Y
    label = label_Placebo,
    fmt = fmt_Placebo
    )

plt.grid() # show gridlines
plt.legend() # Show legend

# Add labels
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')

# Save the Figure
tumor_response = plt
plt.savefig("tumor_response.png")

# Show the Figure
tumor_response.show()
```

```python

# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count_data = data_complete.groupby(['Drug', 'Timepoint'])['Mouse ID'].nunique()

# Convert to DataFrame
df_mice_count = pd.DataFrame(mice_count_data)
df_mice_count.reset_index(inplace=True)

# Preview DataFrame
df_mice_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_mice_count_data_new = df_mice_count.pivot(index='Timepoint',columns='Drug',values='Mouse ID')

# Preview the Data Frame
df_mice_count_data_new
```

```python
# Calculate percentage of mice survival
mice_count = 25

# Function to calculate percentage passing the drug name and list
def get_percentage(drug_name, percentage_list):
    percentage_list = []
    for ind in df_mice_count_data_new.index:
        percentage = round(((df_mice_count_data_new[drug_name][ind])/mice_count) * 100)
        percentage_list.append(percentage)
    return percentage_list

# Percentage of mice survival for Capomulin
mice_count_percentage_capomulin = []
mice_count_percentage_capomulin = get_percentage('Capomulin', mice_count_percentage_capomulin)

# Percentage of mice survival for Infubinol
mice_count_percentage_infubinol = []
mice_count_percentage_infubinol = get_percentage('Infubinol', mice_count_percentage_capomulin)

# Percentage of mice survival for Ketapril
mice_count_percentage_ketapril = []
mice_count_percentage_ketapril = get_percentage('Ketapril', mice_count_percentage_capomulin)

# Percentage of mice survival for Placebo
mice_count_percentage_placebo = []
mice_count_percentage_placebo = get_percentage('Placebo', mice_count_percentage_capomulin)
```

```python
# Generate the Plot (with Error Bars) - Survival During Treatment
y_errors = 0.5
line_width = 1

plt.errorbar(
    x_axis, # X
    mice_count_percentage_capomulin,  # Y
    yerr = y_errors,     # Y-errors
    label = label_Capomulin,
    fmt = fmt_Capomulin,
    linewidth = line_width # width of plot line
    )

plt.errorbar(
    x_axis, # X
    mice_count_percentage_infubinol,  # Y
    label = label_Infubinol,
    fmt = fmt_Infubinol
    )

plt.errorbar(
    x_axis, # X
    mice_count_percentage_ketapril,  # Y
    label = label_Ketapril,
    fmt = fmt_Ketapril
    )

plt.errorbar(
    x_axis, # X
    mice_count_percentage_placebo,  # Y
    label = label_Placebo,
    fmt = fmt_Placebo
    )

plt.grid() # show gridlines
plt.legend() # Show legend

# Add labels
plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')

# Save the Figure
survival_rate = plt
plt.savefig("survival_rate.png")

# Show the Figure
survival_rate.show()
```

```python
# Calculate the percent changes for each drug
# Formula = (drug value at day 45 - drug value a day 0)/45  * 100

duration = 45
percent_change = []

for drug_name in df_mean_tumor_vol_data_new.columns:
    percentage = (((df_mean_tumor_vol_data_new[drug_name].iloc[-1] - df_mean_tumor_vol_data_new[drug_name][0])/duration) * 100)
    percent_change.append({'Drug': drug_name, 'Percentage': percentage})

df_percent_change = pd.DataFrame(percent_change)
```

```python
# Display the data to confirm
df_percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
# get Capomulin, Infubinol, Ketapril, Placebo
#percent_change_tuple = [tuple(x) for x in df_percent_change.to_records(index=False)]
percent_change = list()
percent_change.append(df_percent_change['Percentage'].iloc[0])
percent_change.append(df_percent_change['Percentage'].iloc[2])
percent_change.append(df_percent_change['Percentage'].iloc[3])
percent_change.append(df_percent_change['Percentage'].iloc[5])
percent_change_tuple = tuple(percent_change)
```

```python
tumor_drugs = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']
df_drug_change = pd.DataFrame()
df_drug_change['drug'] = tumor_drugs 
df_drug_change['tumor change'] = percent_change_tuple
```

```python
# Splice the data between passing and failing drugs
# add a new column for a boolean 'passing'
df_drug_change['passing'] = df_drug_change['tumor change'] < 0
df_drug_change
```

```python
# Orient widths. Add labels, tick marks, etc.
x_position = [i for i, _ in enumerate(tumor_drugs)]
tumor_change = plt.bar(
    x_position, 
    df_drug_change['tumor change'], 
    color = df_drug_change.passing.map({True: 'g', False: 'r'}), # green for pass, red for fail
    width = 1,
    edgecolor = 'k' # black
    )
plt.title('Tumor Change Over 45 Day Treatment')
plt.ylabel('% Tumor Volume Change')
plt.grid(linewidth=0.5)
plt.xticks(x_position, tumor_drugs)


# Use functions to label the percentages of changes
def add_label(bar, rects, labels):
    for rect, label in zip(rects, labels):
        y = 2
        if rect.get_facecolor()==(0.0, 0.5, 0.0, 1.0): # set y to -6 if bar is color green
            y = -6            
        plt.text(rect.get_x() + rect.get_width() / 2, 
                 y, 
                 "{:.0f}%".format(label),
                 ha= 'center',
                 va = 'bottom',
                 color = 'white'
                )

# Call functions to implement the function calls
add_label(tumor_change, tumor_change.patches, round(df_drug_change['tumor change']))


# Save the Figure
plt.savefig("tumor_change_bar.png")

# Show the Figure
plt.show()
```
